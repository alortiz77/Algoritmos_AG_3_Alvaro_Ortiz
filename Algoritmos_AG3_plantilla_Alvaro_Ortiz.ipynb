#Funcionas basicas


```python
###############################################################################
# Funciones básicas refactorizadas para mayor eficiencia
###############################################################################

# Se genera una solucion aleatoria con comienzo en el nodo 0
def crear_solucion(Nodos):
    # Optimización: En lugar de restar conjuntos iterativamente,
    # separo el nodo inicial y permuto el resto. Es mucho más rápido (O(N)).
    solucion = [Nodos[0]]
    resto_nodos = Nodos[1:].copy()
    random.shuffle(resto_nodos)
    return solucion + resto_nodos

# Devuelve la distancia entre dos nodos
def distancia(a, b, problem):
    return problem.get_weight(a, b)

# Devuelve la distancia total de una trayectoria/solucion (función objetivo)
def distancia_total(solucion, problem):
    dist_total = 0
    n = len(solucion)
    for i in range(n - 1):
        dist_total += distancia(solucion[i], solucion[i+1], problem)
    # es importante sumar la distancia de retorno al punto de origen para cerrar el ciclo
    dist_total += distancia(solucion[n-1], solucion[0], problem)
    return dist_total

# Prueba de la generación de una solución temporal
sol_temporal = crear_solucion(Nodos)

print("Ruta aleatoria generada:")
print(sol_temporal)
print(f"\nDistancia total de la ruta: {distancia_total(sol_temporal, problem)}")
```

## 1. Búsqueda Aleatoria
Este algoritmo genera soluciones de forma estocástica y conserva la que tenga la menor distancia. Aunque no es un método eficiente para encontrar el óptimo global en problemas de alta dimensionalidad como el TSP, se puede usar como una **línea base** para evaluar el rendimiento de las heurísticas más avanzadas.

```python
###############################################################################
# IMPLEMENTACIÓN: BÚSQUEDA ALEATORIA
###############################################################################

def busqueda_aleatoria(problem, iteraciones):
    """
    Realiza una búsqueda aleatoria pura sobre el espacio de soluciones del TSP.
    """
    nodos = list(problem.get_nodes())

    mejor_solucion = []
    # Inicializo con infinito para asegurar que la primera solución se acepte
    mejor_distancia = float('inf')

    for _ in range(iteraciones):
        # Genero una ruta usando la función que optimizamos previamente
        solucion_actual = crear_solucion(nodos)
        distancia_actual = distancia_total(solucion_actual, problem)

        # Criterio de aceptación: si es estrictamente mejor, la guardo
        if distancia_actual < mejor_distancia:
            mejor_solucion = solucion_actual
            mejor_distancia = distancia_actual

    # Formateo limpio de la salida en consola
    print(f"--- Resultados Búsqueda Aleatoria ({iteraciones} iteraciones) ---")
    print(f"Mejor distancia encontrada : {mejor_distancia}")
    print(f"Ruta                       : {mejor_solucion}\n")

    return mejor_solucion, mejor_distancia

# Ejecuto el algoritmo.
# Uso 10,000 iteraciones ya que la función de generación es rápida.
solucion_aleatoria, dist_aleatoria = busqueda_aleatoria(problem, 10000)
```

## 2. Búsqueda Local con Entornos Variables (VNS)
En este caso en vez de usar una única estructura de vecindario, se puede implementar un esquema **VNS (Variable Neighborhood Search)**.
Primero defino dos entornos:
1. **Swap:** Intercambio de dos nodos individuales.
2. **2-opt:** Inversión de un segmento completo de la ruta (deshace cruces en el grafo).

Entonces, el algoritmo explora el vecindario más simple (Swap). Cuando llega a un mínimo local, cambia dinámicamente al entorno más complejo (2-opt) para intentar escapar del mínimo.

```python
###############################################################################
# GENERADORES DE ENTORNOS (VNS)
###############################################################################

def genera_vecina_swap(solucion, problem):
    """ Entorno 1: Intercambio de posición de 2 nodos """
    mejor_solucion = solucion.copy()
    mejor_distancia = float('inf')
    n = len(solucion)

    # El nodo 0 es el origen, se itera desde el nodo 1
    for i in range(1, n - 1):
        for j in range(i + 1, n):
            vecina = solucion.copy()
            # Swap rápido en Python
            vecina[i], vecina[j] = vecina[j], vecina[i]

            dist = distancia_total(vecina, problem)
            if dist < mejor_distancia:
                mejor_distancia = dist
                mejor_solucion = vecina

    return mejor_solucion, mejor_distancia

def genera_vecina_2opt(solucion, problem):
    """ Entorno 2: Inversión de un segmento (2-opt real) """
    mejor_solucion = solucion.copy()
    mejor_distancia = float('inf')
    n = len(solucion)

    for i in range(1, n - 1):
        for j in range(i + 1, n):
            vecina = solucion.copy()
            # Para invertir el segmento de la ruta entre i y j
            vecina[i:j+1] = reversed(vecina[i:j+1])

            dist = distancia_total(vecina, problem)
            if dist < mejor_distancia:
                mejor_distancia = dist
                mejor_solucion = vecina

    return mejor_solucion, mejor_distancia
```

```python
###############################################################################
# BÚSQUEDA LOCAL (VNS)
###############################################################################

def busqueda_local_vns(solucion_inicial, problem):
    solucion_actual = solucion_inicial.copy()
    distancia_actual = distancia_total(solucion_actual, problem)

    # Defino la jerarquía de entornos (de menos a más disruptivo)
    entornos = [genera_vecina_swap, genera_vecina_2opt]
    k = 0 # Índice del entorno actual
    iteracion = 0

    print("--- Inicio Búsqueda Local con Entornos Variables (VNS) ---")
    print(f"Distancia inicial de referencia: {distancia_actual}")

    # Mientras no haya agotado todos los entornos sin mejora
    while k < len(entornos):
        iteracion += 1

        # Se obtiene la mejor vecina del entorno k actual
        mejor_vecina, dist_vecina = entornos[k](solucion_actual, problem)

        if dist_vecina < distancia_actual:
            # Actualizo y regreso al entorno más básico (k=0)
            solucion_actual = mejor_vecina
            distancia_actual = dist_vecina
            k = 0
        else:
            # Mínimo local alcanzado para el entorno k.
            # Se cambia al siguiente entorno (k+1) para intentar escapar.
            k += 1

    print(f"\nBúsqueda finalizada tras {iteracion} iteraciones (valles locales evaluados).")
    print(f"Mejor distancia final (VNS): {distancia_actual}")
    print(f"Ruta óptima local            : {solucion_actual}\n")

    return solucion_actual, distancia_actual

# Ejecuto el algoritmo VNS tomando la mejor solución de la búsqueda aleatoria
solucion_vns, dist_vns = busqueda_local_vns(solucion_aleatoria, problem)
```

## 3. Recocido Simulado (Simulated Annealing)
El recocido simulado es una metaheurística inspirada en la metalurgia que permite escapar de óptimos locales aceptando soluciones peores con una probabilidad que disminuye gradualmente (Temperatura).

En este caso intente optimizar la función `genera_vecina_aleatorio()`. Inicialmente se usaba una concatenación de listas ineficiente para hacer un *swap*. Para esta version implemente un **2-opt aleatorio real** invirtiendo el segmento entre dos índices elegidos al azar usando *slices* de memoria directos. Esto me permite deshacer cruces en la ruta y acelera la convergencia drásticamente.

```python
###############################################################################
# FUNCIONES AUXILIARES: RECOCIDO SIMULADO
###############################################################################

def genera_vecina_aleatorio(solucion):
    """
    Genera una vecina aleatoria usando un 2-opt real optimizado.
    Invertimos el subsegmento en lugar de hacer un simple intercambio de extremos.
    """
    # Se eligen dos índices aleatoriamente (excluyendo el nodo 0 que es nuestro origen fijo)
    i, j = sorted(random.sample(range(1, len(solucion)), 2))

    vecina = solucion.copy()
    # Se invierte el segmento de la ruta in-place (mucho más rápido que concatenar)
    vecina[i:j+1] = reversed(vecina[i:j+1])

    return vecina

def probabilidad(T, delta):
    """
    Criterio de aceptación de Boltzmann.
    Evalúa si se acepta una solución peor basándose en la temperatura actual.
    """
    return random.random() < math.exp(-delta / T)

def bajar_temperatura(T):
    """ Tasa de enfriamiento geométrico """
    return T * 0.99
```

```python
###############################################################################
# ALGORITMO PRINCIPAL: RECOCIDO SIMULADO
###############################################################################

def recocido_simulado(problem, TEMPERATURA_INICIAL):
    # Primero genero una solución aleatoria de partida
    solucion_actual = crear_solucion(list(problem.get_nodes()))
    distancia_actual = distancia_total(solucion_actual, problem)

    # Inicializo las variables para rastrear el óptimo global histórico
    mejor_solucion = solucion_actual.copy()
    mejor_distancia = distancia_actual

    T = TEMPERATURA_INICIAL
    iteraciones = 0

    print(f"--- Inicio Recocido Simulado (T_inicial = {T}) ---")
    print(f"Distancia inicial aleatoria: {distancia_actual}")

    # Criterio de parada: temperatura residual cercana a cero
    while T > 0.0001:
        iteraciones += 1

        # 1. Generar nueva vecina
        vecina = genera_vecina_aleatorio(solucion_actual)
        distancia_vecina = distancia_total(vecina, problem)

        # Delta: diferencia de distancias (positiva = la vecina es peor)
        delta = distancia_vecina - distancia_actual

        # 2. Criterio de aceptación
        # Se acepta si es mejor (delta < 0), o si la probabilidad térmica lo permite
        if delta < 0 or probabilidad(T, delta):
            solucion_actual = vecina.copy()
            distancia_actual = distancia_vecina

        # 3. Guardar el óptimo global histórico
        if distancia_actual < mejor_distancia:
            mejor_solucion = solucion_actual.copy()
            mejor_distancia = distancia_actual

        # 4. Enfriar el sistema
        T = bajar_temperatura(T)

    print(f"\nBúsqueda finalizada tras {iteraciones} iteraciones térmicas.")
    print(f"Mejor distancia final (SA) : {mejor_distancia}")
    print(f"Ruta óptima                : {mejor_solucion}\n")

    return mejor_solucion, mejor_distancia

# Ejecuto el recocido simulado con una temperatura inicial alta
solucion_sa, dist_sa = recocido_simulado(problem, 10000000)
```

## 4. Visualización de Resultados
Para representar gráficamente el problema del Agente Viajero (TSP), puedo utilizar la técnica de **Escalado Multidimensional (MDS)**. Esto me permite proyectar la matriz de distancias (pesos) en un plano bidimensional, aproximando las coordenadas espaciales de los nodos para poder dibujar el grafo.

Multidimensional scaling problem(MDS):   https://en.wikipedia.org/wiki/Multidimensional_scaling

```python
import warnings
# Ocultamos las advertencias de futuras versiones de sklearn para mantener el notebook limpio
warnings.filterwarnings("ignore", category=FutureWarning)

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

def plot_tsp_solution(distance_matrix, tsp_solution, titulo="Grafo TSP"):
    """
    Dibuja el grafo de un TSP con las posiciones calculadas mediante MDS y muestra
    solo las aristas correspondientes a la solución del TSP.
    """
    # Crear el grafo completo
    G = nx.Graph()
    num_nodes = len(distance_matrix)
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):
            G.add_edge(i, j, weight=distance_matrix[i][j])

    # Usar MDS para calcular posiciones de los nodos
    # Añadi n_init=4 e init='random' para cumplir con las nuevas normativas de sklearn y evitar future warnings
    mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42, n_init=4)
    positions = mds.fit_transform(distance_matrix)

    # Convertir las posiciones en un diccionario para networkx
    pos = {i: positions[i] for i in range(num_nodes)}

    # Crear un subgrafo con las aristas del camino TSP
    TSP_G = nx.Graph()
    for i in range(len(tsp_solution) - 1):
        u = tsp_solution[i]
        v = tsp_solution[i + 1]
        TSP_G.add_edge(u, v, weight=distance_matrix[u][v])

    # Cerrar el ciclo conectando el último nodo con el primero
    TSP_G.add_edge(tsp_solution[-1], tsp_solution[0], weight=distance_matrix[tsp_solution[-1]][tsp_solution[0]])

    # Dibujar el grafo
    plt.figure(figsize=(10, 8))

    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)

    # Dibujar las aristas del camino TSP
    nx.draw_networkx_edges(TSP_G, pos, edge_color='red', width=2)

    # Añadir etiquetas a los nodos y pesos de las aristas
    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    plt.title(titulo)
    plt.show()
```

```python
# Gráfica de la solución generada por Búsqueda Aleatoria (Baseline)
plot_tsp_solution(
    problem.edge_weights,
    solucion_aleatoria,
    titulo=f"Solución Búsqueda Aleatoria (Distancia: {dist_aleatoria})"
)
```

```python
# Gráfica de la solución optimizada por Recocido Simulado (o VNS)
plot_tsp_solution(
    problem.edge_weights,
    solucion_sa,
    titulo=f"Solución Óptima - Recocido Simulado (Distancia: {dist_sa})"
)
```
